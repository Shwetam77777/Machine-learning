```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
dataset = pd.read_csv(r"D:\ML DATASETS\dummy_variable_file.csv")
df = pd.DataFrame(dataset)
df
```

```python
town_dummy=pd.get_dummies(df["town"])
type(town_dummy)
```

```python
town_dummy
```

```python
town_dummy.drop(columns=['west windsor'],axis=1,inplace=True)
town_dummy
```

```python
merged_df=pd.concat([town_dummy,df],axis=1)
```

```python
merged_df
```

```python
merged_df.drop('town',axis=1,inplace=True)
```

```python
X = merged_df.iloc[:,:3]
X.head()
```
