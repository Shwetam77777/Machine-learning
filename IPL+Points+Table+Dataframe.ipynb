```python
new_ipl17=ipl17.loc[:,['Team','Matches','Won']]
new_ipl17
```

```python
new_ipl18=ipl18.loc[:,['Team','Matches','Won']]
new_ipl18
```

```python
merged_dataframe=pd.merge(left=new_ipl17,right=new_ipl18,how='outer',on='Team')
```

```python
#merged_dataframe['Matches_x']+merged_dataframe['Matches_y']
merged_dataframe.fillna(value=0.0,axis=1,inplace=True)
merged_dataframe
```

```python
merged_dataframe['Total_Matches']=merged_dataframe['Matches_x']+merged_dataframe['Matches_y']
merged_dataframe['Total_Won']=merged_dataframe['Won_x']+merged_dataframe['Won_y']
```

```python
merged_dataframe.drop(labels=['Matches_x','Matches_y','Won_x','Won_y'],axis=1,inplace=True)
```

```python
merged_dataframe['win_percentage'] = merged_dataframe['Total_Won']/merged_dataframe['Total_Matches']
merged_dataframe.sort_values(by=['Total_Won','win_percentage'],ascending=False)
```
